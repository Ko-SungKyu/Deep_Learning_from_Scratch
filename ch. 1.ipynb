## 배열의 원소 접근

```python
import numpy as np
```

```python
#배열 생성
x = np.array([[51,55],[14,19],[0,4]])
x
```

```python
#배열 평탄화(1차원 배열로 전환)
x = x.flatten()
x
```

```python
#인덱스를 통한 접근
x[np.array([0,2,4])]
```

```python
#조건식을 통한 접근
x[x>15]
```

## matplotlib

```python
import matplotlib.pyplot as plt
```

```python
# sin 함수의 생성
x = np.arange(0,6,0.1) # 0부터 6까지 0.1간격
y = np.sin(x)

plt.plot(x,y)
plt.show()
```

```python
# sin, cos 함수의 생성
# pyplot 추가기능

x = np.arange(0,6,0.1) # 0부터 6까지 0.1간격
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x,y1, label='sin') # 첫번째 곡선
plt.plot(x,y2, label='cos', linestyle='--') # 두번째 곡선, 점선
plt.xlabel('x') # x축 이름
plt.ylabel('y') # y축 이름
plt.title('sin&cos')# 그래프 타이틀
plt.legend() # 범례


plt.show()
```
